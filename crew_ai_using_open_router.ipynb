```python
import os
from crewai import Agent, Task, Crew, Process
from crewai.llm import LLM
# from crewai_tools import BaseTool # This import is causing the error
from crewai.tools import tool
import requests
import json
import yfinance as yf
from datetime import datetime
from typing import List

# Hard-coded OpenRouter API key
API_KEY = "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"

# ‚úÖ CORRECT: Configure CrewAI's LLM class for OpenRouter
llm = LLM(
    model="openai/gpt-4o-mini",  # Use a working model
    api_key=API_KEY,
    base_url="https://openrouter.ai/api/v1"
)

# Alternative: Use environment variables (CrewAI's preferred method)
os.environ["OPENAI_API_KEY"] = API_KEY
os.environ["OPENAI_API_BASE"] = "https://openrouter.ai/api/v1"  # Use OPENAI_API_BASE instead
os.environ["OPENAI_MODEL_NAME"] = "openai/gpt-4o-mini"

print("‚úÖ LLM configuration fixed!")
```

```python
@tool("Web Data Scraper")
def web_scraping_tool(keyword: str) -> str:
    """Scrapes data from multiple web sources based on keywords."""
    scraped_data = {}

    try:
        # Financial Data using yfinance
        if len(keyword) <= 5 and keyword.isalpha():
            try:
                ticker = yf.Ticker(keyword.upper())
                hist = ticker.history(period="1mo")
                if not hist.empty:
                    scraped_data["financial"] = {
                        "type": "stock_data",
                        "symbol": keyword.upper(),
                        "current_price": float(hist['Close'][-1]),
                        "monthly_data": {
                            "dates": [str(date.date()) for date in hist.index][-10:],  # Last 10 days
                            "prices": [float(price) for price in hist['Close']][-10:],
                            "volumes": [int(vol) for vol in hist['Volume']][-10:]
                        },
                        "price_change": float(hist['Close'][-1] - hist['Close'][0]),
                        "percentage_change": float(((hist['Close'][-1] - hist['Close'][0]) / hist['Close'][0]) * 100)
                    }
                    print(f"‚úÖ Financial data scraped for {keyword.upper()}")
                else:
                    scraped_data["financial"] = {"error": "No financial data found"}
            except Exception as e:
                scraped_data["financial"] = {"error": f"Financial scraping failed: {str(e)}"}

        # News Data (mock structure)
        scraped_data["news"] = {
            "type": "news_data",
            "keyword": keyword,
            "articles": [
                {
                    "title": f"Latest {keyword} market analysis",
                    "summary": f"Current market trends and analysis for {keyword}",
                    "sentiment": "positive",
                    "date": str(datetime.now().date())
                },
                {
                    "title": f"{keyword} quarterly performance review",
                    "summary": f"Detailed quarterly performance analysis for {keyword}",
                    "sentiment": "neutral",
                    "date": str(datetime.now().date())
                }
            ],
            "total_articles": 2,
            "sentiment_summary": {"positive": 1, "neutral": 1, "negative": 0}
        }

        # General Data from Wikipedia
        try:
            wiki_url = f"https://en.wikipedia.org/api/rest_v1/page/summary/{keyword.replace(' ', '_')}"
            response = requests.get(wiki_url, timeout=10)
            if response.status_code == 200:
                data = response.json()
                scraped_data["general"] = {
                    "type": "general_info",
                    "title": data.get("title", keyword),
                    "summary": data.get("extract", f"Information about {keyword}"),
                    "word_count": len(data.get("extract", "").split())
                }
                print(f"‚úÖ General data scraped from Wikipedia for {keyword}")
            else:
                scraped_data["general"] = {
                    "type": "general_info",
                    "title": keyword,
                    "summary": f"General information about {keyword}",
                    "word_count": 50
                }
        except Exception as e:
            scraped_data["general"] = {"error": f"General scraping failed: {str(e)}"}

        return json.dumps(scraped_data, indent=2)

    except Exception as e:
        return f"Scraping error: {str(e)}"

@tool("Data Analyzer")
def data_analysis_tool(raw_data: str) -> str:
    """Analyzes scraped data and extracts insights."""
    try:
        data = json.loads(raw_data)
        analysis_results = {
            "analysis_timestamp": str(datetime.now()),
            "data_sources": list(data.keys()),
            "insights": [],
            "key_metrics": {}
        }

        # Analyze financial data
        if "financial" in data and data["financial"].get("type") == "stock_data":
            financial = data["financial"]
            percentage_change = financial.get("percentage_change", 0)
            current_price = financial.get("current_price", 0)

            if percentage_change > 5:
                analysis_results["insights"].append("üü¢ Strong positive price movement detected (+5% or more)")
            elif percentage_change < -5:
                analysis_results["insights"].append("üî¥ Significant price decline observed (-5% or more)")
            else:
                analysis_results["insights"].append("üü° Moderate price fluctuation within normal range")

            analysis_results["key_metrics"]["stock_performance"] = {
                "current_price": current_price,
                "percentage_change": percentage_change,
                "trend": "bullish" if percentage_change > 0 else "bearish"
            }

        # Analyze news sentiment
        if "news" in data:
            sentiment_summary = data["news"].get("sentiment_summary", {})
            total_articles = sum(sentiment_summary.values())
            if total_articles > 0:
                dominant_sentiment = max(sentiment_summary, key=sentiment_summary.get)
                analysis_results["insights"].append(f"üì∞ News sentiment: {dominant_sentiment} ({sentiment_summary[dominant_sentiment]}/{total_articles} articles)")
                analysis_results["key_metrics"]["news_sentiment"] = dominant_sentiment

        # Analyze general information
        if "general" in data:
            word_count = data["general"].get("word_count", 0)
            if word_count > 100:
                analysis_results["insights"].append("üìö Rich information content available (100+ words)")
            elif word_count > 50:
                analysis_results["insights"].append("üìñ Moderate information content available (50-100 words)")
            else:
                analysis_results["insights"].append("üìÑ Basic information content available (<50 words)")

        analysis_results["summary"] = {
            "total_insights": len(analysis_results["insights"]),
            "data_quality": "Excellent" if len(analysis_results["insights"]) > 2 else "Good",
            "analysis_complete": True
        }

        print(f"‚úÖ Analysis complete - {len(analysis_results['insights'])} insights generated")
        return json.dumps(analysis_results, indent=2)

    except Exception as e:
        return f"Analysis error: {str(e)}"

print("üõ†Ô∏è Tools ready!")
```

```python
def create_scraper_agent():
    return Agent(
        role='Data Scraping Specialist',
        goal='Scrape relevant data from multiple web sources efficiently',
        backstory="""You are an expert data collector with extensive experience in
        gathering information from financial APIs, news sources, and general web content.
        You know how to extract the most relevant and current data for analysis.""",
        verbose=True,
        allow_delegation=False,
        llm=llm,  # ‚úÖ Using the correctly configured LLM
        tools=[web_scraping_tool]
    )

def create_analyst_agent():
    return Agent(
        role='Data Analysis Expert',
        goal='Analyze scraped data and extract meaningful insights and patterns',
        backstory="""You are a seasoned data analyst who excels at identifying trends,
        patterns, and actionable insights from raw data. You have expertise in financial
        analysis, sentiment analysis, and statistical interpretation.""",
        verbose=True,
        allow_delegation=False,
        llm=llm,  # ‚úÖ Using the correctly configured LLM
        tools=[data_analysis_tool]
    )

def create_visualizer_agent():
    return Agent(
        role='Visualization Specialist',
        goal='Plan and recommend the best visualizations for analyzed data',
        backstory="""You are a data visualization expert who knows how to present
        complex data insights through clear, compelling charts and graphs. You understand
        which visualization types work best for different data patterns.""",
        verbose=True,
        allow_delegation=False,
        llm=llm  # ‚úÖ Using the correctly configured LLM
    )

print("ü§ñ Agents configured with working LLM!")
```

```python
def working_data_analysis(keyword: str):
    """Working data analysis with proper error handling."""
    print(f"üîç Starting analysis for: {keyword}")
    print("=" * 50)

    # Create agents with fixed LLM
    scraper = create_scraper_agent()
    analyst = create_analyst_agent()
    visualizer = create_visualizer_agent()

    # Create tasks with clearer descriptions
    scrape_task = Task(
        description=f"""
        Scrape comprehensive data about '{keyword}' from multiple sources.

        Focus on:
        1. Financial data (if applicable) - stock prices, market trends
        2. Recent news articles and sentiment
        3. General information from reliable sources

        Use the web_scraping_tool to gather data and return structured JSON results.
        """,
        agent=scraper,
        expected_output="JSON formatted data containing financial, news, and general information about the keyword"
    )

    analyze_task = Task(
        description=f"""
        Analyze the scraped data about '{keyword}' to extract key insights.

        Your analysis should:
        1. Identify important trends and patterns
        2. Assess financial performance if stock data is available
        3. Evaluate news sentiment and public perception
        4. Summarize key findings and metrics

        Use the data_analysis_tool to process the scraped data.
        """,
        agent=analyst,
        expected_output="Comprehensive analysis with insights, trends, and key metrics extracted from the data"
    )

    visualize_task = Task(
        description=f"""
        Based on the analysis results for '{keyword}', recommend the best visualization approaches.

        Consider:
        1. Stock price charts if financial data is available
        2. Sentiment analysis visualizations for news data
        3. Key metrics dashboard layouts
        4. Trend identification charts

        Provide specific recommendations for chart types, layouts, and styling.
        """,
        agent=visualizer,
        expected_output="Detailed visualization recommendations including chart types, layouts, and design specifications"
    )

    # Create crew with proper configuration
    crew = Crew(
        agents=[scraper, analyst, visualizer],
        tasks=[scrape_task, analyze_task, visualize_task],
        process=Process.sequential,
        verbose=True
    )

    try:
        print("üöÄ Starting crew execution...")
        result = crew.kickoff()

        print("\n" + "=" * 50)
        print("‚úÖ ANALYSIS COMPLETED SUCCESSFULLY!")
        print("=" * 50)

        return result

    except Exception as e:
        print(f"\n‚ùå Analysis Error: {str(e)}")
        print("Check your API key and model availability")
        return None

# Test the working system
print("üéØ Testing fixed system...")
result = working_data_analysis("AAPL")

if result:
    print(f"\nüìä Final Result:\n{result}")
else:
    print("\n‚ö†Ô∏è  Analysis failed - check error messages above")
```
